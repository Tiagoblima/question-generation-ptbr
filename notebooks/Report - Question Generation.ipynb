```python
import pandas as pd
import json
import os

BASE_DIR = "/content/drive/MyDrive/QuestionGeneration/Reports/tiagoblima"

results = []
for exp in os.listdir(BASE_DIR):

  scores_path = os.path.join(BASE_DIR, exp, "scores.json")
  with open(scores_path) as f:
    scores = json.load(f)

  exp_name = exp.replace("qg-", "")
  scores.update({
      "model": exp_name
  })
  results.append(scores)

df = pd.DataFrame(results)
df.set_index('model', inplace=True)
df.sort_index(inplace=True)
df.to_csv()
```
